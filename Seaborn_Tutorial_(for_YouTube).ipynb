```python
from tensorflow.keras.utils import get_file
from copy import deepcopy
import os

zip_path = get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)

csv_path, _ = os.path.splitext(zip_path)

weather_df = pd.read_csv(csv_path)

weather_df
```

```python
weather_df = weather_df[::6*24][:365]
weather_df.reset_index(drop=True, inplace=True)

weather_df_full = deepcopy(weather_df)

weather_df = weather_df[['Date Time', 'p (mbar)', 'T (degC)']]

weather_df
```

```python
weather_df['Date Time']
```

```python
weather_df = deepcopy(weather_df)
weather_df['Date Time'] = pd.to_datetime(weather_df['Date Time'],
                                         format='%d.%m.%Y %H:%M:%S')

weather_df['Date Time']
```

```python
weather_df.head()
```

```python
sns.lineplot(x='Date Time', y='T (degC)', data=weather_df)
```

```python
g = sns.lineplot(x='Date Time', y='T (degC)', data=weather_df)
g.figure.autofmt_xdate()
```

```python
weather_df_shuffled = weather_df.sample(frac=1)
weather_df_shuffled.reset_index(drop=True, inplace=True)

weather_df_shuffled
```

```python
g = sns.lineplot(x='Date Time', y='T (degC)', data=weather_df_shuffled)
g.figure.autofmt_xdate()
```

```python
weather_df['above_average_pressure'] = (weather_df['p (mbar)'] > weather_df['p (mbar)'].mean()).astype(int)
weather_df
```

```python
g = sns.lineplot(x='Date Time', y='T (degC)', markers=True, style='above_average_pressure',
                 hue='above_average_pressure',
                 data=weather_df)
g.figure.autofmt_xdate()
```

```python
weather_df['above_average_pressure'].value_counts()
```

```python
g = sns.relplot(x='Date Time', y='T (degC)', markers=True,
                 style='above_average_pressure',
                 hue='above_average_pressure',
                 col='above_average_pressure',
                 kind='line',
                 data=weather_df)
g.figure.autofmt_xdate()
```
